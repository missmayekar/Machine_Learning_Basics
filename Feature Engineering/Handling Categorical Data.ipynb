# 2. K-1 OneHotEncoding
To solve the problem of multi-collinearity we drop one column each from the one hot encoded features, usually first one for easy understanding.

```python
pd.get_dummies(df, columns=['brand', 'fuel', 'owner'], drop_first=True)
```

# 3. One Hot Encoding using scikit learn
Pandas doesn't remember the order in which the features were encoded so the order can change the next time.
Generally scikit learn is used in ML for one hot encoding

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(df.iloc[:,0:4],df. iloc[:, -1], test_size=0.2, random_state = 42)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop='first', sparse_output=False, dtype=np.int32)
```

```python
X_train_new = ohe.fit_transform(X_train[['brand','fuel','owner']])
```

```python
X_train_new
```

```python
X_test_new = ohe.transform(X_test[['brand','fuel','owner']])
```
