```python
tips = sns.load_dataset( 'tips')
```

```python
tips.sample(10)
```

```python
titanic = pd.read_csv('Titanic-Dataset.csv')
```

```python
flights = sns.load_dataset( 'flights')
```

```python
flights.sample(10)
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.sample(10)
```

## Numerical to Numerical Analysis

### 1. Scatterplot

```python
# Bivariate analysis between two numerical features
# Every point signifies a customer
sns.scatterplot(x=tips['total_bill'], y=tips['tip'])
plt.title('Scatter Plot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
# Multivariate analysis of two numerical and one categorical feature
sns.scatterplot(x=tips['total_bill'], y=tips['tip'], hue=tips['sex'], style=tips['smoker'], size=tips['size'])
plt.title('Scatter Plot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

### 2. Pair plot

```python
# Multivariate analysis of multiple numerical features
sns.pairplot(iris, hue='species')
```

### 3. Line plot

```python
new = flights.groupby('year')['passengers'].sum().reset_index()
new
```

```python
sns.lineplot(x = new['year'], y = new['passengers'])
```

```python
sns.heatmap(flights.pivot_table(values='passengers', index= 'month', columns= 'year'))
```

## Numerical - Categorical

### 2. Bar Plot

```python
# Bivariate analysis of a numerical and a categorical column
# x axis is generally category and y axis is numerical
sns.barplot(x = titanic['Pclass'], y = titanic['Age'])
```

```python
# Multivariate analysis of one numerical and two categorical feature
sns.barplot(x = titanic['Pclass'], y = titanic['Fare'], hue=titanic['Sex'])
```

### 3. Box Plot


```python
# Bivariate analysis between one numerical and one categorical
sns.boxplot(x = titanic['Sex'], y = titanic ['Age'])
```

```python
# Multivariate analysis between one numerical and two categorical
sns.boxplot(x = titanic['Sex'], y = titanic[ 'Age'], hue=titanic ['Survived'])
```

### 4. Histplot 

```python
sns.histplot(titanic[titanic['Survived']==0][ 'Age'], kde=True)
sns.histplot(titanic[titanic ['Survived']==1][ 'Age'], kde = True)
```

```python
# if only kde is needed
sns.kdeplot(titanic[titanic['Survived']==0][ 'Age'], label='Not Survived')
sns.kdeplot(titanic[titanic ['Survived']==1][ 'Age'], label='Survived')
plt.legend()
```

## Categorical - Categorical

### 1. Heat Map

```python
# Lighter colour indicates a higher number - it can be seen below that large number of people died in Pclass 1
# While a large number of people survived in Pclass 1
sns.heatmap(pd.crosstab(titanic[ 'Pclass'],titanic[ 'Survived']))
```

### 2. Cluster Map

```python
sns.clustermap(pd.crosstab(titanic[ 'Parch'],titanic['Survived']))
```
